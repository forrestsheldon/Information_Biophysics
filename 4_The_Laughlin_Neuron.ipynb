Large Monopoloar Cells (LMCs) in the insect eye are interneurons that give graded responses to inputs in their visual field.  Lateral inhibition and their intensity response makes them sensitive to fluctuations in relative, rather than absolute intensity. These cells pass information to later stages of visual processing and as such, ideas from information theory may be relevant to their coding.  We will find that maximizing the mutual information between the input and output predicts the coding strategy of the neuron and that this lead the output of the neuron to look indistinguishable from white noise.  This means the filter should just be the CDF of the input distribution!

Suppose the output of the neuron is a deterministic function of the input $r = g(s)$ and that the input is drawn from the distribution $p(s)$.  Maximising the mutual information $I(s, r) = H(r) - H(r|s) = H(r)$ amounts to maximizing the entropy of the distribution of $r$ as the conditional entropy for a deterministic function is zero.  Consider outputs that lie on the interval $[r_{min}, r_{max}]$ such that the target density of the output is $p(r) = \frac{1}{r_{max} - r_{min}}$.

The probability mass on the signal interval $[s,s+\Delta s]$ is $p(s)\Delta s$.  This is mapped to the interval $g(s+\Delta s) - g(s)$ of the ouput.  In order to maintain normalization, we thus require,
$$p(s)\Delta s = p(r)\Delta r = \frac{1}{r_{max} - r_{min}}(g(s+\Delta s) - g(s)).$$
This leads to,
$$\frac{(g(s+\Delta s) - g(s))}{\Delta s} = (r_{max} - r_{min}) p(s)$$
Requiring that $g$ be continuous and differentiable, in the limit $\Delta s \to 0$ is
$$g'(s) = (r_{max} - r_{min}) p(s).$$
from which we get
$$g(s) = (r_{max} - r_{min}) \int_{s_{min}}^s p(s') ds'$$
so the filter is just the CDF of the input!  This has been confirmed by exposing the retinas to various signal distributions and observing the corresponding filters.  It is found that the filters are adapted to the signal distribution of the environment.

### Adding Noise

I am handling this in the continuous case so it may be worth reviewing the continuous entropy notebook before jumping in.

We now add some gaussian noise to our coding function, $r = g(s) + \xi$.  We assume that $\langle \xi\rangle = 0$ and $\langle \xi^2\rangle = \sigma_\xi^2(s)$ so that the variance of the noise may be a function of the signal. In this case, we will find that the filter should allocate an interval to the signal $s$ proportional to $\frac{1}{\sigma_\xi(s)}$.

To show this, we know that the conditional distribution $p(r|s)$ is gaussian centered at $g(s)$,
$$ p(r|s) = \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(r - g(s))^2}{2\sigma^2}}.$$  This leads to a conditional entropy of,
$$\ln 2\,H(r|s) = \frac{1}{2}\int\, ds \,p_s(s) \ln (2\pi e \sigma^2(s))$$
and thus a mutual information,
$$\ln 2\, I(r,s) = -\int p_r(r) \ln p_r(r) - \frac{1}{2}\int\, ds \,p_s(s) \ln (2\pi e \sigma^2(s)).$$
To find the optimal distribution, we want to vary this with respect to $p_r$, but to do that we need to know the dependence of $p_s$ on $p_r$.  To find this, we consider $p_r(r) = \int ds p(r|s)p_s(s)$ and expand in the limit of low noise,
$$p_r(r) = \int ds \frac{1}{\sqrt{2\pi \sigma^2}} e^{-\frac{(r - g(s))^2}{2\sigma^2}} p_s(s).$$
If $\sigma$ is small, the gaussian restricts the integral to the area close to $y=g(s)$, so we can expand the neighboring function in the integral and get an expansion in powers of $\sigma$!  To this end, we change variables to $g(s) = z$,
$$p_r(r) = \int dz \frac{1}{\sqrt{2\pi \sigma^2\left(g^{-1}(z)\right)}} e^{-\frac{(r - z)^2}{2\sigma^2}} \frac{p_s\left(g^{-1}(z)\right)}{g'(g^{-1}(z)}.$$
We now expand $f(z) = \frac{p_s\left(g^{-1}(z)\right)}{g'(g^{-1}(z)}$ about $z=y$, discarding all odd order terms as they will integrate to zero with the gaussian,
$$\frac{p_s\left(g^{-1}(z)\right)}{g'(g^{-1}(z)} = \left.\frac{p_s\left(g^{-1}(z)\right)}{g'(g^{-1}(z)}\right|_{z=r} + \\
\frac{(z-r)^2}{2!}\left.\left[ \frac{p_s''(g^{-1}(z))}{g'(g^{-1}(z))^3} - 
3 \frac{p_s'(g^{-1}(z))}{g'(g^{-1}(z))^4}g''(g^{-1}(z)) + 
3 \frac{p_s(g^{-1}(z))}{g'(g^{-1}(z))^4}g''(g^{-1}(z))^2 +
\frac{p_s(g^{-1}(z))}{g'(g^{-1}(z))^4}g'''(g^{-1}(z))\right]\right|_{z=r}  + ...$$
Performing the integral yields,
$$\frac{p_s\left(g^{-1}(z)\right)}{g'(g^{-1}(z)} = \frac{p_s\left(g^{-1}(r)\right)}{g'(g^{-1}(r)} +\\ 
\frac{\sigma^2}{2}\left.\left[ \frac{p_s''(g^{-1}(z))}{g'(g^{-1}(z))^3} - 
3 \frac{p_s'(g^{-1}(z))}{g'(g^{-1}(z))^4}g''(g^{-1}(z)) + 
3 \frac{p_s(g^{-1}(z))}{g'(g^{-1}(z))^4}g''(g^{-1}(z))^2 +
\frac{p_s(g^{-1}(z))}{g'(g^{-1}(z))^4}g'''(g^{-1}(z))\right]\right|_{z=r}  + ...$$
So corrections to the deterministic case come in at $\sigma^2$. Using the 0th order approximation, (which is the same as taking the deterministic transformation $r = g(s)$ giving $p_s(s)ds = p_r(r)dr$), we transform the second integral in $I$ to depend on $p_r$) and add the normalization constraint
$$\ln 2\, I(r,s) = -\int p_r(r) \ln p_r(r) - \frac{1}{2}\int\, dr \,p_r(r) \ln (2\pi e \sigma^2(r)) + \lambda (\int dr \,p_r(r) - 1).$$
Where $\sigma$ has been similarly transformed to depend on $r$.  This gives the variation,
$\frac{\delta I}{\delta p_r} = -\ln p_r(r) - 1 - \frac{1}{2}\ln 2\pi e\sigma^2 + \lambda.$
Solving for $p_r(r)$ and grouping constants, we find that the dependence on the noise is,
$p_r(r) = \frac{1}{Z}\frac{1}{\sigma(r)}$ which is our result.  In the event of uniform noise, we get a uniform output.  If the noise grows with $s$, we would expect the output distribution to be skewed towards smaller $s$.  Note that in this derivation, we have restricted $\sigma$ to be small but have put no restriction on it's dependence on $s$.
