A number of biological systems (e.g. drosophila embryo, dictyostelium) synchronize their movements with signals that far outpace diffusion.  How?  Here we examine a Turing-like mechanism utilizing reaction-diffusion phenomena.  We first note that in normal diffusion/linear elliptic equations, traveling wave solutions are not possible:

We consider a traveling wave to be a solution of the form $\phi(x - ct)$.  Subsitution into the diffusion equation yields the ODE,
$$-c\phi' = D\phi''$$
where derivates are with respect to the argument $\phi(z)$. This leads to the solution $\phi(z) = A + Be^{-cz/D}$ which cannot be made finite for all $x$ and $t$ and thus is not a physical solution.

In contrast, reaction-diffusion equations can exhibit solutions of a traveling wave type.

### Intuition

Consider a single chemical species that is produced at a constant rate, experiences decay, and also self-promotes it's own production up to some peak.  Examined as a function of species concentration, this will lead to the following curve

Notice that this curve gives three fixed points, two of which ($\phi_{LOW}$, $\phi_{HIGH}$) are stable and one which is unstable $\phi_U$.  In the absence of diffusion (strong mixing, or a single point system) the system will move towards one of the stable points.  If we consider initial conditions of the system in a step configuration as displayed below, diffusion will attempt to smooth this step but will be countered by the reaction term.  If the rates at which the reaction term pull the concentration towards $\phi_L$ and $\phi_H$ are different, this will result in a shift of the step function, ie a traveling wave front.  A similar phenomenon could be imagined for a fluctuation that exceeds some threshold.

The previous examples give some useful intuition.  Traveling wave solutions should generally look like a translating step, or spreading pulse whose speed should be linked to the diffusivity and reaction rates.  In particular, it should depend on some imbalance between how the system moves toward each.  Let's try to formalize this a bit.

### A Mechanical Analogy

We wish to understand under what circumstances the RDE can exhibit wave solutions.  To this end, we assume the basic form of $R(\phi)$ given above with boundary conditions $\phi(\infty) = \phi_L,\,\phi(-\infty) = \phi_H$ and use some intuition from mechanics to understand what this should tell us about the traveling waves and when they can occur.

We begin by substituting our wave ansatz $\phi(x - ct)$ into the RDE $\partial_t \phi = R(\phi) + D \partial^2_x \phi$, yielding the ODE,
$$-c \phi' = R(\phi) + D\phi''$$
where a $'$ denotes differentiation with respect to the argument of $\phi(z)$.
Rearranging some terms we note a close similarity to the equations of motion in mechanics subject to linear damping,
$$ D\phi'' = -R(\phi) - c\phi' \qquad mx'' = -\partial_x V(x) - b x'$$
To complete the analogy we consider a fixed point $x$ and run time backwards so that the argument of $\phi$ is also increasing.  In this case, as the wave front above travels backwards from right to left, $\phi$ moves from $\phi_H$ to $\phi_L$.  Consider the potential energy given by our 'forcing' term $R$,

```python
Image(filename='BioWaves_R_Potential.jpeg')
```

Our solution $\phi$ must lead to the particle rolling down the hill and stopping at $\phi_L$ which can only occur if the 'dissipation' from our $c\phi'$ compensates for the difference in energies of the two states:
$$ V(\phi_H) - V(\phi_L) = \int_{\phi_L}^{\phi_H} d\phi \,R(\phi) = - \int_{\phi_L}^{\phi_H} d\phi\, c\phi'$$
This constraint may be used to find the speed $c$ at which the wave propogates.  In particular, we confirm that if the two lobes of $R$ integrate to zero, no motion of the wave can occur.  As the RHS is positive for our waveform, the integral of $R$ will decide the direction of propogation of the wave.

### The Reduced Nagumo Model/Schlogl Model

As a soluble example, we take for $R(\phi)$ the cubic,
$$ R(\phi) = r(\phi - \phi_L)(\phi_U - \phi)(\phi - \phi_H)$$
As our solution ansatz, we take the differential equation $\phi' = \alpha(\phi - \phi_L)(\phi - \phi_H)$ which is appropriately exponential near the stable points.  Plugging this into our ODE from the wave ansatz and canceling factors gives,
$$-c\alpha = r(\phi_U - \phi) + D\alpha^2 (2\phi - \phi_H - \phi_L).$$
Rearranging,
$$D\alpha^2(\phi_H + \phi_L) - c\alpha - r\phi_U = (2D\alpha^2 - r)\phi.$$
In order for this to hold for all values of $\phi$ both sides of the equation must vanish.  This gives
$$\alpha = \sqrt{\frac{r}{2D}}\\
c = \sqrt{\frac{Dr}{2}}(\phi_H + \phi_L - 2\phi_U).$$
As our intuition hinted, the wave speed increases with the diffusivity and the magnitude of the asymmetry in the reaction.  Here the reaction will progress, and thus the wave will move, towards whichever stable fixed point is further from the unstable point.
